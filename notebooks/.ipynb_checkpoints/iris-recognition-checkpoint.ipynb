```python
import pandas as pd 
import numpy as np
import os
import matplotlib.pyplot as plt
datapath=os.getcwd()
datadir='/'.join(datapath.split('/')[:-1])+'/data'
datadir
dir=os.getcwd().split('/')
datadir='/'.join(dir[0:-1])+'/data'
modeldir='/'.join(dir[0:-1])+'/models'
```

```python
np.random.seed(42)
df=pd.read_csv(datadir+'/iris.csv')
df.head()
df=df.sample(frac=1)
```

```python
df_stetosa=df.loc[df["Species"]=="setosa"]
df_versicolor=df.loc[df["Species"]=="versicolor"]
df_virginica=df.loc[df["Species"]=="virginica"]
fig, ((ax1,ax2,ax3),(ax4,ax5,ax6))=plt.subplots(2,3,figsize=(16,20))

ax1.scatter(x=df_stetosa.index,y=df_stetosa["Sepal.Length"] )
ax1.scatter(x=df_stetosa.index,y=df_stetosa["Petal.Length"])
ax1.set(title='setosa lenght')
ax1.legend(["Sepal","Petal"])

ax4.scatter(x=df_stetosa.index,y=df_stetosa["Sepal.Width"] )
ax4.scatter(x=df_stetosa.index,y=df_stetosa["Petal.Width"])
ax4.set(title='setosa width')
ax4.legend(["Sepal","Petal"])

ax2.scatter(x=df_versicolor.index,y=df_versicolor["Sepal.Length"] )
ax2.scatter(x=df_versicolor.index,y=df_versicolor["Petal.Length"])
ax2.set(title='versicolor lenght')
ax2.legend(["Sepal","Petal"])

ax5.scatter(x=df_versicolor.index,y=df_versicolor["Sepal.Width"] )
ax5.scatter(x=df_versicolor.index,y=df_versicolor["Petal.Width"])
ax5.set(title='versicolor width')
ax5.legend(["Sepal","Petal"])

ax3.scatter(x=df_virginica.index,y=df_virginica["Sepal.Length"] )
ax3.scatter(x=df_virginica.index,y=df_virginica["Petal.Length"])
ax3.set(title='virginica lenght')
ax3.legend(["Sepal","Petal"])

ax6.scatter(x=df_virginica.index,y=df_virginica["Sepal.Width"] )
ax6.scatter(x=df_virginica.index,y=df_virginica["Petal.Width"])
ax6.set(title='virginica width')
ax6.legend(["Sepal","Petal"])
```

```python
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.calibration import calibration_curve

x=df.drop("Species",axis=1)
y=df["Species"]
def set_number(x):
    if x=='setosa':
        return 0
    elif (x=="versicolor"):
        return 1
    else:
        return 2
y=y.apply(lambda x: set_number(x))
y

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
model=RandomForestClassifier(n_estimators=20)
model.fit(x_train,y_train)
y_preds=model.predict(x_test)
print(str(model.score(x_test,y_test)))

plt.scatter([i for i in range(len(y_preds))],y_preds)
plt.scatter([i for i in range(len(y_test))],y_test)
ax=plt.figure()
```

```python
import pickle
pickle.dump(model,open(modeldir+'/iris_recognition_model.pkl','wb'))
```
